```python
# -*- coding: utf-8 -*-
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import numpy as np

# detector_backend options: "opencv", "ssd", "dlib", "mtcnn", "retinaface"
# Using the updated extract_faces function instead of deprecated detectFace

# Extract faces from the image
faces = DeepFace.extract_faces(img_path="dataset/testing/modi1.jpg",
                                detector_backend="opencv",
                                enforce_detection=False)

# Get the first detected face
if faces and len(faces) > 0:
    # The face is returned as a numpy array in RGB format
    face_detected = faces[0]['face']
    
    # Normalize pixel values if needed (extract_faces returns normalized values 0-1)
    if face_detected.max() <= 1.0:
        face_detected = (face_detected * 255).astype(np.uint8)
    
    # Display using matplotlib (Jupyter notebook friendly)
    plt.figure(figsize=(6, 6))
    plt.imshow(face_detected)
    plt.axis('off')
    plt.title('Detected Face')
    plt.show()
else:
    print("No face detected in the image")
```
