```python
#importing the required libraries
import cv2
import face_recognition
import matplotlib.pyplot as plt

#loading the image to detect
image_to_detect = cv2.imread('images/testing/trump-modi.jpg')

#detect all faces in the image
#arguments are image,no_of_times_to_upsample, model
all_face_locations = face_recognition.face_locations(image_to_detect,model='hog') #model='cnn' better but need GPU support

#print the number of faces detected
print(f"There are {len(all_face_locations)} no of faces in this image")

#looping through the face locations
for index,current_face_location in enumerate(all_face_locations):
    #splitting the tuple to get the four position values of current face
    top_pos,right_pos,bottom_pos,left_pos = current_face_location
    #printing the location of current face
    print(f"Found face {index+1} at top:{top_pos},right:{right_pos},bottom:{bottom_pos},left:{left_pos}")
    #slicing the current face from main image
    current_face_image = image_to_detect[top_pos:bottom_pos,left_pos:right_pos]
    #showing the current face with dynamic title
    #  convert BGR  RGB for correct color display in matplotlib
    current_face_image_rgb = cv2.cvtColor(current_face_image, cv2.COLOR_BGR2RGB)

    # showing the current face with dynamic title in Jupyter
    plt.figure(figsize=(4, 4))
    plt.imshow(current_face_image_rgb)
    plt.title("Face # " + str(index + 1))
    plt.axis("off")
    plt.show()



```
